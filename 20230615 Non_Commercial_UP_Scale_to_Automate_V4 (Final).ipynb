```python
# Folder path containing the files
pathSpecified = "D:/Shaikh_Desktop_Mydoc/Desktop/Automation Projects/Non Commercial/All Non COmmercial reports"  # Replace with the actual folder path


# Import os module  
import os  
# Define the specified path  
  
# Using listdir() function  
listOfFileNames = os.listdir(pathSpecified)


# Print the name of all files in directory  

print("Following is the list of names of all the files present in the specified directory: ")  
print(listOfFileNames)
```

```python
import pandas as pd

# Folder path containing the files
folder_path = "D:/Shaikh_Desktop_Mydoc/Desktop/Automation Projects/Non Commercial/All Non Commercial reports"

# Create an empty DataFrame to store the combined data
combined_data = pd.DataFrame()

# Iterate over the file names
for file_name in listOfFileNames:
    # Construct the file path
    file_path = os.path.join(folder_path, file_name)
    # Read the file into a DataFrame
    df = pd.read_excel(file_path, header=None, index_col=False, dtype={0: str, 8: str, 9: str})
    df["BizNFac"]= df.iloc[0,0]
    df[16] = pd.to_datetime(df[16], errors='coerce')
    df[18] = pd.to_datetime(df[18], errors='coerce')
    df['Source'] = file_name
    

#     df = df.iloc[2:].reset_index(drop=True)
        
#     df.columns = df.iloc[0]
    
    # Append the DataFrame to the combined_data DataFrame
    combined_data = combined_data.append(df)



# Print the combined data
combined_data.head(6)
```

```python
combined_data.iloc[0:3, [-1]]= "Source"
combined_data.iloc[0:3, [-2]] = "BizNFac"



# data = combined_data.iloc[4:]
# combined_data_col = combined_data.iloc[2]
# combined_data_A =  pd.DataFrame(data, columns=combined_data_col)
```

```python
combined_data.columns = combined_data.iloc[2]
combined_data = combined_data.iloc[3:].reset_index(drop=True)
combined_data.head(4)
```

```python
# Get the current column labels
columns = combined_data.columns


for i, col in enumerate(columns):
    if pd.isna(col):
        new_col = f"Column{i+1}"
        combined_data.rename(columns={col: new_col}, inplace=True)
```

```python
# Rename the columns
MonthlyData = combined_data.rename(columns={'Receive': 'Receive Date', 
                                          'Receive.1': 'Receive Quantity', 
                                          'QC Received': 'QC Received On' ,
                                         'VAT Chalan' : 'VAT Chalan No', 
                                          'Vat Chalan' : 'Vat Chalan Date'})


MonthlyData.columns
```

```python
# Iterate over each row in reverse order
for i in range(len(MonthlyData) - 1, 0, -1):
    if pd.isnull(MonthlyData.loc[i, 'Code']):
        # Append the values to the immediate previous row
        prev_name = str(MonthlyData.loc[i-1, 'Name'])
        current_name = str(MonthlyData.loc[i, 'Name'])
        MonthlyData.loc[i-1, 'Name'] = prev_name + ' ' + current_name
        
        # Drop the current row
        MonthlyData.drop(i, inplace=True)

# Reset the index of the DataFrame
MonthlyData.reset_index(drop=True, inplace=True)

# Print the updated DataFrame
MonthlyData.head(4)
```

```python
MonthlyData.columns
```

* Trim data

```python
#Remove nan from the Name column

MonthlyData['Name'] = MonthlyData['Name'].str.replace(' nan', '')

# Trim the columns "Code", "Name", and "UMO"
MonthlyData['Code'] = MonthlyData['Code'].str.strip()
MonthlyData['Name'] = MonthlyData['Name'].str.strip()
MonthlyData['UoM'] = MonthlyData['UoM'].str.strip()
```

#### Output Data

```python
# Create a Pandas Excel writer using XlsxWriter as the engine.
# path = 
writer = pd.ExcelWriter(r'C:\Users\shaikh.kamran\ProjectPurchase\OutPutFiles\NonCom_20230910_02.xlsx', engine='xlsxwriter')

# Write the MonthlyData DataFrame to a worksheet.
MonthlyData.to_excel(writer, sheet_name='NonCom', index=False)

# Save the writer and close it.
writer.save()
writer.close()
```
