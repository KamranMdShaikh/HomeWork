* Average Rate

```python
# Concatenate 'Year' and 'Month' into a new column 'Concatenated_Column'
combined_data_22['YYYY-MMM'] = combined_data_22['Year'].astype(str) + '-' + combined_data_22['Month']
display(combined_data_22.head(5))
display(combined_data_22['YYYY-MMM'].unique())
display(combined_data_22.shape)



combined_data_AVRate =combined_data_22.copy()

dates_to_keep_WAP_AVRate= ['2023-January', '2023-February','2023-March', '2023-April', '2023-May', '2023-June']


combined_data_WAP3 = combined_data_AVRate[combined_data_AVRate['YYYY-MMM'].isin(dates_to_keep_WAP_AVRate)]

display(combined_data_WAP3['YYYY-MMM'].unique())


# Group by 'Item Code', 'Year', and 'Month', and calculate the WAP
AVRate_2021_2022 = combined_data_WAP3.groupby([ 'Business', 'Item Code', 'Currency'])['PORate'].agg('mean')

# Reset the index and rename the column
AVRate_2021_2022 = AVRate_2021_2022.reset_index().rename(columns={'PORate': 'AVRate'})

# Print the resulting grouped data
display(AVRate_2021_2022.head(5))
```

* WAP  WAP_fy_2021_2022

```python
# Concatenate 'Year' and 'Month' into a new column 'Concatenated_Column'
combined_data_22['YYYY-MMM'] = combined_data_22['Year'].astype(str) + '-' + combined_data_22['Month']
display(combined_data_22.head(5))
display(combined_data_22['YYYY-MMM'].unique())
display(combined_data_22.shape)



combined_data_WAP2 =combined_data_22.copy()


combined_data_WAP = combined_data_WAP2[combined_data_WAP2['YYYY-MMM'].isin(dates_to_keep_WAP_AVRate)]

display(combined_data_WAP['YYYY-MMM'].unique())


def calculate_wap(amount, quantity):
    if quantity != 0:
        wap = amount / quantity
        return wap
    else:
        return None

# Group by 'Item Code', 'Year', and 'Month', and calculate the WAP
WAP_fy_2021_2022 = combined_data_WAP.groupby(['Business', 'Item Code', 'Currency']).apply(lambda x: calculate_wap(x['Amount'].sum(), x['POQuantity'].sum()))

# Reset the index and rename the column
WAP_fy_2021_2022 = WAP_fy_2021_2022.reset_index().rename(columns={0: 'WAP'})

# Print the resulting grouped data
display(WAP_fy_2021_2022.head(5))
```
